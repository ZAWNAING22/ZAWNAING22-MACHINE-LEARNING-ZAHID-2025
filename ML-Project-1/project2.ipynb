```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score

#laod data
data=pd.read_csv("student_performance_dataset.csv")
#input and ouput
X=data[['Study_Hours_per_Week']]
y=data['Final_Exam_Score']

#train model
model=LinearRegression()
model.fit(X,y)
predicted_scores=model.predict(X)
df=pd.DataFrame({
    'predicted_scores':predicted_scores,
    'Acutual_score':data['Final_Exam_Score']
})
df.head(2)
```

```python
#Model Evaluation
mae=mean_absolute_error(y,predicted_scores)
mse=mean_squared_error(y,predicted_scores)
rmse=np.sqrt(mse)
r2=r2_score(y,predicted_scores)

print("MAE:",round(mae,2))
print("MSE:",round(mse,2))
print("RMSE",round(rmse,2))
print("r2_score(Model accuracy):",round(r2))
```

```python
#histogram
plt.figure(figsize=(10,5))
plt.hist(data['Final_Exam_Score'],bins=30, color='skyblue', edgecolor='Black')
plt.title("Distrubution of final Exam score:")
plt.xlabel("Final Exam Score")
plt.ylabel("Number of student")
plt.grid(True)
plt.show()
```

```python
#scatter + Regression Line
plt.Figure(figsize=(10,6)) #width and height
plt.scatter(X,y, color="Blue", label="Actual Scores")
plt.plot(X,predicted_scores,color="Red", label="Predicted Score(Regression line)")
plt.title("Model Precdiction Vs actual score:")
plt.xlabel("Study Hourse Per week")
plt.ylabel("Final Output")
plt.grid(True)
plt.show()

hours=25
predicted_new_scores=model.predict([[hours]])
print(f"the predicted score based on weekly study hour{hours} is :{int(predicted_new_scores)}")
plt.savefig("Model prediction vs atucal score.png")
```
